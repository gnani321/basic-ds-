```python
# Standardization (Z-score normalization):

# Preferred when:
# 1. Features have different units or scales:  Standardization transforms data to have zero mean and unit variance. This makes features with vastly different scales comparable, which is crucial for many machine learning algorithms (e.g., k-NN, SVM, linear regression) that are sensitive to feature scaling.
# 2. Algorithms assume normally distributed data: Some algorithms, such as linear discriminant analysis, work best when the data is normally distributed.  Standardization can help to approximate a normal distribution, although it does not guarantee normality.
# 3. Outliers are present: While standardization can be affected by outliers, it does not remove them and still provides a consistent rescaling across the dataset.  Robust scaler is preferred when significant outliers are present.



# Min-Max Scaling (Normalization):

# Preferred when:
# 1. Data is not normally distributed: If your data is not normally distributed or the assumption of normality is not crucial, min-max scaling is an excellent choice as it transforms data to the range [0, 1].
# 2. Feature scaling is required for algorithms that are not sensitive to the distribution: Some algorithms, like decision trees, are relatively insensitive to feature scaling, but normalizing data to a specific range can still be beneficial in terms of computational efficiency and improving the model's performance.
# 3. Interpretability of the scaled features:  Min-max scaling provides a more intuitive interpretation of the scaled data as it preserves the relative relationships of the original data points in the new scale.
# 4. When working with neural networks:  Neural networks often perform better with input features in the [0, 1] range or other restricted ranges, thus normalization is frequently used.

```

# Encoding Techniques:

```python
#•	Apply One-Hot Encoding to categorical variables with less than 5 categories.
for col in categorial_col:
    if data[col].nunique() < 5:
        one_hot_encoded = pd.get_dummies(data[col], prefix=col)
        data = pd.concat([data, one_hot_encoded], axis=1)
```

```python
#•	Use Label Encoding for categorical variables with more than 5 categories.
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for col in categorial_col:
    if data[col].nunique() > 5:
        data[col] = label_encoder.fit_transform(data[col])
```

```python
#•	Discuss the pros and cons of One-Hot Encoding and Label Encoding.
# One-Hot Encoding avoids imposing an ordinal relationship where there is none, but it can lead to a high-dimensional dataset.
# Label Encoding is compact but introduces ordinal relationships where none might exist (which can be problematic for some algorithms).
```

# Feature Engineering:

```python
# creating new featueres
data['age-education_num']=data['age']*data['education_num']
data['capital_gain_ratio']=data['capital_gain']/(data['capital_loss']+1)
data.head()
```

```python
#log transformation one skewed feature
import numpy as np
data['age_log']=np.log1p(data['age'])
data.head()
```

# Feature Selection:

```python
# isloation forest
from sklearn.ensemble import IsolationForest
iso_forest=IsolationForest(contamination=0.05,random_state=42)
outliers=iso_forest.fit_predict(data[numerical_col])
outliers
data['outliers']=outliers
data['outliers'].value_counts()
data[data['outliers']==-1]

data=data.drop('outliers',axis=1)
data.head()
```

```python
pps_matrix=pps.matrix(data)
pps_matrix
```

```python
# display the heatmap of pps matrix
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,10))
sns.heatmap(pps_matrix.pivot(index='x',columns='y',values='ppscore'),annot=True,fmt=".2f",cmap='viridis')
plt.title('Predictive Power Score Matrix')
plt.show()
```

```python
#correlation
correlation_matrix=data.corr()
plt.figure(figsize=(12,10))
sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```


*  Conclusion on Feature Selection:
*  PPS score helps identify predictive relationships that might not be captured by Pearson correlation. For instance, some features that show weak or no correlation may have predictive power, especially for non-linear relationships.
